```python
import tensorflow as tf
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('Untitled Document 1.csv',names=['l1','l2','l3','l4','l5','l6','l7'])
```

```python
data_copy = data
```

```python
p = np.asarray([1,0])
a = np.asarray([0,1])
data['l7'] = data['l7'].map({1:p,2:a})
```

```python
data['l1'] = data['l1']/103
data['l2'] = data['l2']/138
data['l3'] = data['l3']/155
data['l4'] = data['l4']/182
data['l5'] = data['l5']/297
data['l6'] = data['l6']/20

```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['l1','l2','l3','l4','l5','l6']]
y = data['l7']
```

```python
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_test
```

```python
learning_rate = 0.27
training_epochs = 15
```

```python
x=tf.placeholder(tf.float32,shape=[None,6])
y_=tf.placeholder(tf.float32,shape=[None, 2])
```

```python
def multilayer_perceptron(x, weights, biases):
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.sigmoid(layer_1)
        
    out_layer = tf.matmul(layer_1, weights['out']) + biases['out']
    out_layer = tf.nn.sigmoid(out_layer)
    return out_layer
```

```python
n_input = 6
n_hidden_1 = 10
n_classes = 2
```

```python
weights = {
    'h1': tf.Variable(tf.random_normal([n_input, n_hidden_1])),
    
    'out': tf.Variable(tf.random_normal([n_hidden_1, n_classes]))
}
```

```python
biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1])),
    
    'out': tf.Variable(tf.random_normal([n_classes]))
}
```

```python
pred = multilayer_perceptron(x, weights, biases)
```

```python
loss = tf.reduce_sum( tf.square( y_ - pred ) )
```

```python
train_step = tf.train.MomentumOptimizer(0.27,0.77).minimize(loss)
correct_prediction = tf.equal(tf.argmax(pred,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sess = tf.InteractiveSession()
init = tf.initialize_all_variables()
sess.run(init)
epoch=2000
```

```python
for step in range(epoch):
   _, c=sess.run([train_step,loss], feed_dict={x: X_train, y_:[t for t in y_train.as_matrix()]})
   if step%500==0 :
       print(c)
```

```python
print(sess.run(accuracy,feed_dict={x: X_test, y_:[t for t in y_test.as_matrix()]})) 
```
